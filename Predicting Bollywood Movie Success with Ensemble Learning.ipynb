```python
# Here, Item_Song(Y/N) feature is represented as a numeric column, but it is a categorical feature. So, we will change the
# data type of this feature to object data type.
```

```python
df['Item_Song(Y/N)'] = df['Item_Song(Y/N)'].astype('object')
```

```python
df.dtypes
```

```python
# get list of only numeric features
num_col = list(set(list(df._get_numeric_data().columns))- {'Success/Faliure'})
```

```python
num_col
```

```python
# Outliers removal
```

```python
# standardize function
def standardize(raw_data):
    return ((raw_data - np.mean(raw_data, axis = 0)) / np.std(raw_data, axis = 0))
```

```python
df.shape
```

```python
df[num_col] = standardize(df[num_col])
import scipy as sp
df = df[(np.abs(sp.stats.zscore(df[num_col])) < 3).all(axis=1)]
```

```python
df.shape
```

```python
# 7 rows got dropped from the dataset.
```

```python
df.head().T
```

```python
df.describe()
```

```python
df.dtypes
```

```python
# Many columns are of type object i.e. strings. These need to be converted to ordinal type.
```

```python
# Decision tree in Python can take only numerical / categorical colums. It cannot take string / obeject types. 
# The following code loops through each column and checks if the column type is object then converts those columns
# into categorical with each distinct value becoming a category or code.

for feature in df.columns: # Loop through all columns in the dataframe
    if df[feature].dtype == 'object': # Only apply for columns with categorical strings
        df[feature] = pd.Categorical(df[feature]).codes # Replace strings with an integer
```

```python
# K-Fold Cross Validation on the Algorithms and Comparison
```

```python
from sklearn import model_selection
```

```python
models = []
models.append(('DecisionTree', dt_model))
models.append(('RandomForest', Rf_model))
models.append(('BaggingClassifier', bgcl))
models.append(('GradientBoostingClassifier', gbcl))
models.append(('XGradientBoostingClassifier', xgbcl))
models.append(('AdaBoostClassifier', abcl))
results = []
names = []
scoring = 'accuracy'
for name, model in models:
	kfold = model_selection.KFold(n_splits=10,random_state=2)
	cv_results = model_selection.cross_val_score(model, X, y, cv=kfold, scoring=scoring)
	results.append(cv_results)
	names.append(name)
	msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
	print(msg)
# boxplot algorithm comparison
fig = plt.figure()
fig.suptitle('Algorithm Comparison')
ax = fig.add_subplot(111)
plt.boxplot(results)
plt.xticks(rotation = 90)
ax.set_xticklabels(names)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import VotingClassifier
from sklearn.model_selection import cross_val_score
```

```python
# Stacking
```

```python
lrcl = LogisticRegression(random_state=1)
rfcl = RandomForestClassifier(n_estimators=150,criterion='entropy',max_depth=5 ,random_state=1)
nbcl = GaussianNB()
bgcl = BaggingClassifier(base_estimator=dt_model, n_estimators=21, max_samples= .7, bootstrap=True, oob_score=True)  
enclf = VotingClassifier(estimators = [('lor', lrcl), ('rf', rfcl), ('nb', nbcl), ('bg', bgcl)], voting = 'hard')
```

```python
for clf, label in zip([lrcl , rfcl, nbcl, enclf, bgcl], ['Logistic Regression', 'RandomForest', 'NaiveBayes', 'Ensemble', 'Bagging']):
    scores = cross_val_score(clf, X, y, cv=10, scoring='accuracy')
    print("Accuracy: %0.02f (+/- %0.2f) [%s]" % (scores.mean(), scores.std(), label ))
```
