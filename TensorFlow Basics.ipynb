### Load tensorflow

```
import tensorflow as tf
```

```
tf.__version__
```

### Create a tensor

```
hello = tf.constant('Hello World')
```

```
print(hello)
```

### Tensor name and shape

```
hello_2 =  tf.constant('Hello World', shape=[6,2], name='my_tensor')
print(hello_2)
```

### Adding two tensors

```
a = tf.constant([1.2])
b = tf.constant([3.4])
c = tf.add(a,b)
print(c)
```

This builds the nodes for the computational graph

### How to get tensor Output

At this point, we have built our graph and now we will execute it.

```
#Start a session
sess = tf.Session()
```

```
#Execute hello
print(sess.run(hello))
```

```
#Execute hello_2
print(sess.run(hello_2))
```

```
#Adding a and b
sess.run(c)
```

Close the session

```
sess.close()
```
